### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f'El ingreso total de la tienda 1 es : ${ingreso_total_tienda:,.2f}.')
print(f'El ingreso total de la tienda 2 es : ${ingreso_total_tienda2:,.2f}.')
print(f'El ingreso total de la tienda 3 es : ${ingreso_total_tienda3:,.2f}.')
print(f'El ingreso total de la tienda 4 es : ${ingreso_total_tienda4:,.2f}.')
```

# 3. Calificación promedio de la tienda


```python
# Tienda 1
calif1 = tienda["Calificación"].mean()
print("Tienda 1 - Promedio:", round(calif1, 2))

# Tienda 2
calif2 = tienda2["Calificación"].mean()
print("Tienda 2 - Promedio:", round(calif2, 2))

# Tienda 3
calif3 = tienda3["Calificación"].mean()
print("Tienda 3 - Promedio:", round(calif3, 2))

# Tienda 4
calif4 = tienda4["Calificación"].mean()
print("Tienda 4 - Promedio:", round(calif4, 2))
```

# 4. Productos más y menos vendidos

```python
# Contamos cuántas veces se vendió cada producto
productos_vendidos = tienda["Producto"].value_counts()
productos_vendidos2 = tienda2["Producto"].value_counts()
productos_vendidos3 = tienda3["Producto"].value_counts()
productos_vendidos4 = tienda4["Producto"].value_counts()

# Producto más vendido
mas_vendido = productos_vendidos.idxmax()
cantidad_mas_vendido = productos_vendidos.max()
#tienda 2
mas_vendido2 = productos_vendidos2.idxmax()
cantidad_mas_vendido2 = productos_vendidos2.max()
#tienda 3
mas_vendido3 = productos_vendidos3.idxmax()
cantidad_mas_vendido3 = productos_vendidos3.max()
#tienda 4
mas_vendido4 = productos_vendidos4.idxmax()
cantidad_mas_vendido4 = productos_vendidos4.max()

# Producto menos vendido
menos_vendido = productos_vendidos.idxmin()
cantidad_menos_vendido = productos_vendidos.min()
#tienda 2
menos_vendido2 = productos_vendidos2.idxmin()
cantidad_menos_vendido2 = productos_vendidos2.min()
#tienda 3
menos_vendido3 = productos_vendidos3.idxmin()
cantidad_menos_vendido3 = productos_vendidos3.min()
#tienda 4
menos_vendido4 = productos_vendidos4.idxmin()
cantidad_menos_vendido4 = productos_vendidos4.min()

# Mostramos los resultados
print("Producto más vendido - Tienda 1:", mas_vendido, "-", cantidad_mas_vendido, "ventas")
print("Producto menos vendido - Tienda 1:", menos_vendido, "-", cantidad_menos_vendido, "ventas")

print("Producto más vendido - Tienda 2:", mas_vendido2, "-", cantidad_mas_vendido2, "ventas")
print("Producto menos vendido - Tienda 2:", menos_vendido2, "-", cantidad_menos_vendido2, "ventas")

print("Producto más vendido - Tienda 3:", mas_vendido3, "-", cantidad_mas_vendido3, "ventas")
print("Producto menos vendido - Tienda 3:", menos_vendido3, "-", cantidad_menos_vendido3, "ventas")

print("Producto más vendido - Tienda 4:", mas_vendido4, "-", cantidad_mas_vendido4, "ventas")
print("Producto menos vendido - Tienda 4:", menos_vendido4, "-", cantidad_menos_vendido4, "ventas")
```

# 5. Envío promedio por tienda

```python
#Calcular el promedio de la columna que tiene los costos de envío para cada tienda, para saber cuánto se gasta en promedio en logística.
print(tienda.columns)
```

```python
#calcular envio
costo_envio = tienda['Costo de envío']
costo_envio2 = tienda2['Costo de envío']
costo_envio3 = tienda3['Costo de envío']
costo_envio4 = tienda4['Costo de envío']

#calcular promedio
promedio_envio = costo_envio.mean()
print('El promedio del envio de la tienda 1: es ',round(promedio_envio, 2))
promedio_envio2 = costo_envio2.mean()
print('El promedio del envio de la tienda 2: es ',round(promedio_envio2, 2))
promedio_envio3 = costo_envio3.mean()
print('El promedio del envio de la tienda 3: es ',round(promedio_envio3, 2))
promedio_envio4 = costo_envio4.mean()
print('El promedio del envio de la tienda 4: es ',round(promedio_envio4, 2))

```
